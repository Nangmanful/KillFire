```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('progress985.csv')
df2 = pd.read_csv('random_train_iteration(INITIAL-FINAL, 985,5.69).csv')

# Smoothing (EMA)
col = 'env_runners/episode_return_mean'
ema_span = 100
ema = df[col].ewm(span=ema_span, adjust=False).mean()

# Plot
plt.figure(figsize=(12, 7))
plt.plot(df[col], alpha=0.2, linewidth=1)
plt.plot(ema, color='purple', label=f'MAPPO (EMA span={ema_span}))', linewidth=2)
plt.plot(df2['iteration'], df2['mean_reward'], color='green', linestyle='-', label='Random', linewidth=2)

plt.title('MAPPO vs Random (m = 985, n =5.69)')
plt.xlabel('Index / Iteration')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
