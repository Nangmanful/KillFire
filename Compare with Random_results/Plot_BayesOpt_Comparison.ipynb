```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('progress985.csv')
df2 = pd.read_csv('random_train_iteration(INITIAL-FINAL, 985,5.69).csv')

# Smoothing (EMA)
col = 'env_runners/episode_return_mean'
ema_span = 100
ema = df[col].ewm(span=ema_span, adjust=False).mean()

# Plot
plt.figure(figsize=(12, 7))
plt.plot(df[col], alpha=0.2, linewidth=1)
plt.plot(ema, color='red', label=f'MAPPO (EMA span={ema_span}))', linewidth=3)
plt.plot(df2['iteration'], df2['mean_reward'], color='green', linestyle='-', label='Random', linewidth=3)

plt.title('MAPPO vs Random (m=985, n=5.69)',fontsize=20)
plt.xlabel('Iteration (Batch size: 400)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.legend(fontsize=14) 
plt.tight_layout()
plt.savefig('MAPPO vs Random (m=985, n=5.69).png', dpi=700)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df2 = pd.read_csv('progress985.csv')
df1 = pd.read_csv('progress 855.csv')
df3 = pd.read_csv('progress941.csv')

# Smoothing (EMA)
col = 'env_runners/episode_return_mean'
ema_span = 100
ema1 = df1[col].ewm(span=ema_span, adjust=False).mean()
ema2 = df2[col].ewm(span=ema_span, adjust=False).mean()
ema3 = df3[col].ewm(span=ema_span, adjust=False).mean()
# Plot
plt.figure(figsize=(12, 7))
plt.plot(df1[col], alpha=0.1, linewidth=1, color='red')
plt.plot(ema1, color='red', label=f'm=855, n=8.37 (EMA span={ema_span}))', linewidth=3)

plt.plot(df2[col], alpha=0.1, linewidth=1, color='blue')
plt.plot(ema2, color='blue', label=f'm=985, n=5.69 (EMA span={ema_span}))', linewidth=3)

plt.plot(df3[col], alpha=0.1, linewidth=1, color='green')
plt.plot(ema3, color='green', label=f'm=941, n=7.96 (EMA span={ema_span}))', linewidth=3)

plt.title('Maximize AUC & Final return',fontsize=20)
plt.xlabel('Iteration (Batch size: 400)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.legend(fontsize=14) 
plt.tight_layout()
plt.savefig('Maximize AUC, Final return', dpi=700)
plt.show()
```
